```python
from scipy.stats import shapiro

columns_to_check = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for column in columns_to_check:
    _, p_value = shapiro(df[column])

    if p_value > 0.05:
        print(f"The data in column '{column}' is normally distributed.")
    else:
        print(f"The data in column '{column}' is not normally distributed.")
```

```python
#replacing 0 values with median of that column
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())#normal distribution
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())#normal distribution
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].median())#skewed distribution
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].median())#skewed distribution
df['BMI']=df['BMI'].replace(0,df['BMI'].median())#skewed distribution
```

Oversampling

```python
#oversampling for training
msk = df_train['Outcome'] == 1
num_to_oversample = len(df_train) - 2*msk.sum() #186

df_positive_oversample = df_train[msk].sample(n=num_to_oversample, replace=True, random_state=100)
df_train_oversample = pd.concat([df_train, df_positive_oversample])
df_train_oversample['Outcome'].value_counts()
```

```python
#making the changes for new data which now balanced using oversampling technique
X_train = df_train_oversample.drop(columns=['Outcome'])
y_train = df_train_oversample['Outcome']
```

```python
# Split the test data
X_test = df_test.drop(columns=['Outcome'])
y_test = df_test['Outcome']
# Split the test data
X_test = df_test.drop(columns=['Outcome'])
y_test = df_test['Outcome']
```

Model Selection

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# Separate features and target variable for training data
X_train = df_train_oversample.drop(columns=['Outcome'])
y_train = df_train_oversample['Outcome']

# Split the test data
X_test = df_test.drop(columns=['Outcome'])
y_test = df_test['Outcome']

# Create a Random Forest classifier object
rf_classifier = RandomForestClassifier()

# Train the Random Forest classifier on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data using the trained model
y_pred = rf_classifier.predict(X_test)
# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the performance metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)


```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report

param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'sigmoid'],
    'gamma': [0.1, 0.01, 0.001]
}

svm_classifier = SVC()
grid_search = GridSearchCV(svm_classifier, param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Parameters:", best_params)

y_pred = grid_search.best_estimator_.predict(X_test)

# print classification report
print(classification_report(y_test, y_pred))
 # Accuracy: 0.7792207792207793
# Precision: 0.6388888888888888
# Recall: 0.8518518518518519
# F1 Score: 0.7301587301587301
```

```python
# lets try Randomised search for svc
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from scipy.stats import reciprocal, uniform

param_dist = {
    'C': reciprocal(0.1, 100),
    'kernel': ['linear', 'rbf', 'sigmoid'],
    'gamma': reciprocal(0.001, 0.1)
}

svm_classifier = SVC()
random_search = RandomizedSearchCV(svm_classifier, param_distributions=param_dist, scoring='accuracy', n_iter=10, cv=10, random_state=100)
random_search.fit(X_train, y_train)

best_params = random_search.best_params_
print("Best Parameters:", best_params)

y_pred = random_search.best_estimator_.predict(X_test)

```
